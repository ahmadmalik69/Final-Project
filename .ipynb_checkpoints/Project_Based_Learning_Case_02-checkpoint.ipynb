# Data Preparation

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import statistics as stats

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Import data -- Silakan di-import berdasarkan direktori masing-masing
house = pd.read_csv('train_house.csv')
```

```python
total = house.isnull().sum().sort_values(ascending=False)
percent = (house.isnull().sum()/house.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
# Drop variabel NaN
house.dropna(axis = 1)
```

```python
# Memisahkan data menjadi numerikal & kategorikal
num_cols = house._get_numeric_data().columns
house_numeric = house[num_cols]
house_categorical = house.drop(num_cols, axis = 1)
```

## Variabel Kategorikal

### Utilities

* `Utilities` berisi informasi mengenai fasilitas apa saja yang tersedia di properti tersebut
		
       AllPub	All public Utilities (E,G,W,& S)	
       NoSeWa	Electricity and Gas Only	

```python
#Utilities
sns.barplot(x=house_categorical['Utilities'].value_counts().index,
                 y=house_categorical['Utilities'].value_counts(normalize = True))
plt.title("Chart Persentase Fasilitas") 
plt.xlabel("Fasilitas")  
plt.ylabel("Persentase")
plt.show()
```

### HouseStyle

HouseStyle: Style properti (misal: 1 lantai, 2 lantai, dst)
	
       1Story	One story
       1.5Fin	One and one-half story: 2nd level finished
       1.5Unf	One and one-half story: 2nd level unfinished
       2Story	Two story
       2.5Fin	Two and one-half story: 2nd level finished
       2.5Unf	Two and one-half story: 2nd level unfinished
       SFoyer	Split Foyer
       SLvl	Split Level

```python
#HouseStyle
sns.barplot(x=house_categorical['HouseStyle'].value_counts().index,
                 y=house_categorical['HouseStyle'].value_counts(normalize = True))
plt.title("Chart Persentase Style Properti") 
plt.xlabel("Style Properti")  
plt.ylabel("Persentase")
plt.show()
```

### ExterQual

ExterQual: Kualitas material eksterior 
		
       Ex	Excellent
       Gd	Good
       TA	Average/Typical
       Fa	Fair

```python
#ExterQual
sns.barplot(x=house_categorical['ExterQual'].value_counts().index,
                 y=house_categorical['ExterQual'].value_counts(normalize = True))
plt.title("Chart Persentase Kualitas Material Eksterior") 
plt.xlabel("Kualitas Material Eksterior")  
plt.ylabel("Persentase")
plt.show()
```

#### **Analisis tentang variabel kategorikal**

* *Dari analisis Utilities, didapatkan nilai hampir sebesar sebesar 100 persen rumah sudah memiliki All public utilities.*
* *Dari analisis HouseStyle, didapatkan hampir 50% rumah merupakan rumah satu lantai, sedangkan hampir 30% lainnya merupakan rumah 2 lantai. HouseStyle lainnya memiliki persentase yang sedikit*
* *Dari analisis ExterQual, didapatkan lebih dari 60% kualtias material eksteriornya rata-rata/average/typical, sementara lebih dari 30% memiliki kualitas 'Good', kualitas 'Excelent' dan 'Fair' dalam persentase yang kecil*

## Variabel Numerikal

### SalePrice

* SalePrice merupakan variabel target yang ingin diprediksi
* Secara fisik, ini berarti harga properti yang dijual

```python
x = house_numeric["SalePrice"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Harga Properti") 
plt.xlabel("Harga Properti")  
plt.ylabel("Densitas")
plt.show()
```

#### **Analisis tentang variabel `SalePrice`**

* *Distribusi berupa right skewed*
* *Kebanyakan data dalam nilai sekitar 100.000 dengan rata rata senilai 180.921, Apakah ini wajar untuk harga rumah ?*

### GrLivArea

* GrLivArea berisi informasi mengenai luas area tinggal di atas permukaan tanah (tidak termasuk basement)
* Satuan yang digunakan adalah *square feet*

```python
x = house_numeric["GrLivArea"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Luas Area Tinggal di Atas Permukaan Tanah") 
plt.xlabel("Luas Area Tinggal di Atas Permukaan Tanah")  
plt.ylabel("Densitas")
plt.show()
```

### GarageArea

* GarageArea berisi informasi mengenai luas garasi (jika ada)
* Satuan yang digunakan adalah *square feet*

```python
house_numeric = house_numeric[(house_numeric.GarageArea > 0)]
```

```python
x = house_numeric["GarageArea"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Luas Garasi") 
plt.xlabel("Luas Garasi")  
plt.ylabel("Densitas")
plt.show()
```

#### **Analisis tentang variabel `GarageArea`**

* *Rata rata luas garasi pada tiap rumah sebesar 472.98. Namun ada rumah yang tidak memiliki garasi*

## Pemrosesan Data Variabel Numerikal

```python
# Transformasi log-10 SalePrice

x = np.log10(house_numeric["SalePrice"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Harga Properti") 
plt.xlabel("Harga Properti")  
plt.ylabel("Densitas")
plt.show()
```

```python
# Transformasi log-10 GrLivArea

x = np.log10(house_numeric["GrLivArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Luas Area Tinggal di Atas Permukaan Tanah") 
plt.xlabel("Luas Area Tinggal di Atas Permukaan Tanah")  
plt.ylabel("Densitas")
plt.show()
```

```python
# Menghilangkan nilai nol pada GarageArea

x = np.log10(house_numeric["GarageArea"])
sns.distplot(x, axlabel=x.name, bins = 30)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1,line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.title("Chart Densitas Luas Garasi") 
plt.xlabel("Luas Garasi")  
plt.ylabel("Densitas")
plt.show()
```

#### **Penjelasan dan analisis tentang pemrosesan data**

* *Setelah dilakukan Transformasi log-10, Saleprice GrLivArea, dan GarageArea menjadi memiliki distribusi yang normal*
